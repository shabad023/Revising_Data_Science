# GroupBy

Allows you to groupby rows based of a column then perform some aggregate function on them

Aggregate fn is just that takes a bunch of values and spits out a single value

```python
data={
    'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
     'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
      'Sales':[200,120,340,124,243,350]
     }
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')['Sales'].sum()

# Company de layi sales nu groupby karna hai

# Matlab ('') wala column show karo then agge [''] eh aala column


# Step 1: df.groupby('Company')

# Group rows having the same company together.

# Google:
# 100
# 200

# Amazon:
# 150
# 250
# 300

# Step 2: ['Sales']

# From each group, select only the Sales column.

# So now pandas sees:

# Google Sales:
# 100
# 200

# Amazon Sales:
# 150
# 250
# 300

# Step 3: .describe()

# Calculate summary statistics for each group.

# Pandas computes:

# * count → number of values
# * mean → average
# * std → standard deviation
# * min → minimum
# * 25% → first quartile
# * 50% → median
# * 75% → third quartile
# * max → maximum

#          count   mean       std    min    25%    50%    75%    max
# Company
# Amazon     3.0  233.3   76.38    150   200    250    275    300
# Google     2.0  150.0   70.71    100   125    150    175    200
```

```python
df.groupby('Company')['Company'].sum()
```

# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='sex')

# estimator value by default is mean.
#average of total_bill per categorical value.
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='sex',estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips,hue='sex')
# number dass dinda. Count of Male and Female
# y axis is already chosen that is the counts value
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='day')

# x is categorical , y is numeric
#vich aali line blue box de = median i.e. 50th percentile
#utte aali 75th percentile
#thalle aali 25th percentile
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
#sex nu hue banao but bina split de ohne 2 violingraph bana dene si
# taan split=True iko reh dinda oh violinplot nu
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
#Jitter naal you could see density a lil better
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
# this plot does not overlaps
#stripplot + violinplot

# Caution- par wadde dataset ch zyada hi wide chale jande aa
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='day')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
# ehte kind ch koi vi plot da naam likhdo oh bann jega
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

# Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tips.corr(numeric_only=True)
```

```python
tc=tips.corr(numeric_only=True)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

```python
flights.pivot_table(index='month',columns='year',values='passengers')
# IITG ch use hoya si aah
#matlab iss graph nu eve reshape kardo ke index=month hoje
#columns=year hoje
#values=passengers hoje
#long format to wide format
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=3)
```

```python
# kisi vi 2 cheezan da corelation pata karan layi heatmap kamm anda
```

```python
sns.clustermap(fp,cmap='coolwarm')
# sab toh similar sabtoh kol kol aa jange
#1952 & 1959 are very similar to each other
# so are May and December
```

```python
sns.heatmap(fp.isnull(),linewidth=1)
# koi missing value nahi hai
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
# legend 0-1 ch change ho gaya
```
